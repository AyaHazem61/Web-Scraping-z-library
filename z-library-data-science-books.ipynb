```python
print('Scraping .....')

with open('z-library-data-science.csv','w', encoding = 'utf-8') as file :
    
    obj = csv.writer(file)
    obj.writerow(['BookName' , 'Publisher' , 'Author' , 'Year','Language'])
    
    for i in range(1,11) :
        
        url = f'https://eg1lib.org/s/data%20science?page={i}'
        con = requests.get(url)
    
        soup = bs(con.content , 'html.parser')
        containers = soup.find_all('div' , class_="resItemBox resItemBoxBooks exactMatch")
        
        for container in containers :
        
            title = container.find_all('h3' , itemprop="name")
            title = title[0].text
        
            try :
                publisher = container.find_all('a',title="Publisher")
                publisher = publisher[0].text
            except :
                publisher = 'nan'
        
            author = container.find_all('div',class_="authors")
            author = author[0].text
        
            year = container.find_all('div',class_= "property_value")
            year = year[0].text
            try :
                lan = container.find_all('div',class_="property_value text-capitalize")
                lan = lan[0].text
            except :
                lan = 'nan'
            
            obj.writerow([title ,publisher ,author ,year ,lan])
            
print('Done!')
        
```
