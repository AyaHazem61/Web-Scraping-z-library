```python
import requests
from bs4 import BeautifulSoup as bs
import csv
import pandas as pd
```

```python
with open('all-pages-z-library.csv','w',encoding = 'utf-8') as file :
    obj = csv.writer(file)
    obj.writerow(['title' , 'publisher' , 'author' , 'year'])
    for i in range(1,11) :
        
        url = f'https://eg1lib.org/s/programming%20concepts/?type=phrase&page={i}'
        request = requests.get(url)
        soup = bs(request.content,'html.parser')
        
        containers = soup.find_all('div', class_ ="resItemBox resItemBoxBooks exactMatch")
        
        for container in containers :
            
            title = container.find_all('h3',itemprop="name")
            title = title[0].text.strip()
            
            try :
                publisher = container.find_all('a',title="Publisher")
                publisher = publisher[0].text
            except :
                publisher = 'nan'
                
            author = container.find_all('div', class_="authors")
            author = author[0].text
            year = container.find_all('div', class_="property_value")
            year = year[0].text
            
            obj.writerow([title , publisher , author , year])
        
```
